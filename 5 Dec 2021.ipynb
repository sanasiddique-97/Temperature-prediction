```python
from sklearn.cluster import KMeans
```

```python
target = df1["Temperature"].to_numpy().reshape(-1,1)
target
```

```python
iner = []
for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit(target)
    iner.append(km.inertia_)
    
```

```python
plt.plot(range(1,11), iner)
plt.scatter(range(1,11), iner, color = "red")
plt.show()
```

```python
km = KMeans(n_clusters = 3)
km.fit(target)
```
